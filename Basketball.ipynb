```python
# Import pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in CSV from https://www.basketball-reference.com/leagues/NBA_2021_per_game.html
df = pd.read_csv("basketballref2020.csv")
```

```python
# Print the first 5 rows
print(df.head())
```

```python
# Get the summary
print(df.describe())
```

```python
# Print just the player column
print(df["Player"])
```

```python
# Create a new dataframe that is a copy of the old one
dfNew = df;
```

```python
# Create a new column in the new dataframe that is just the player's name (without the 9 character ID code)
dfNew['Name'] = dfNew['Player'].str[:-10]
print(dfNew.iloc[20:21, 20:])
print(dfNew.info())
```

```python
# Create a new dataframe that reorders the columns in the old dataframe and only includes 9 defensive stats
dfReordered = dfNew[['Rk', 'Name', 'Age', 'Tm', 'ORB', 'DRB', 'TRB', 'STL', 'BLK']]
print(dfReordered.head())
```

```python
# See a list of the column names
cols = list(dfReordered.columns.values)
print(cols)
```

```python
dfReordered.mean(axis=0, skipna = True)
```

```python
fig, axs = plt.subplots(nrows=1)

sns.distplot(
    dfReordered['Age']
)
```

```python
fig, axs = plt.subplots(nrows=1)

sns.stripplot(
    dfReordered['DRB'],
    jitter=0.2, 
)
```

```python
sns.boxplot(
    dfReordered['BLK'],
    width = 0.3
)
```

```python
fig, axs = plt.subplots(nrows=2)

sns.distplot(
    dfReordered['Age'], 
    ax=axs[0],
    color='green',
)
sns.stripplot(
    dfReordered['Age'], 
    jitter=0.4, 
    size=3,
    ax=axs[1],
    alpha=0.3,
    color='green'
)
age = sns.boxplot(
    dfReordered['Age'],
    width=0.3, 
    ax=axs[1],
    showfliers=False,
)

# Align axes
spacer = np.max(dfReordered['Age']) * 0.05
xmin = np.min(dfReordered['Age']) - spacer
xmax = np.max(dfReordered['Age']) + spacer
axs[0].set_xlim((xmin, xmax))
axs[1].set_xlim((xmin, xmax))

# Remove some axis text
axs[0].xaxis.set_visible(False)
axs[0].yaxis.set_visible(False)
axs[1].yaxis.set_visible(False)

# Put the two plots together
plt.subplots_adjust(hspace=0)

# Adjust boxplot fill to be white
axs[1].artists[0].set_facecolor('white')

# Rename Axis
age.set_xlabel("Age", fontsize = 18)
```

```python
joint1 = sns.jointplot(
    x='ORB', 
    y='BLK', 
    data=dfReordered,
    kind="reg",
    ratio=4,
    space=0,
    scatter_kws={
        's': 3,
        'alpha': 0.25
    },
);

joint1.ax_joint.set_xlabel('Offensive Rebounds', fontsize = 18)
joint1.ax_joint.set_ylabel('Blocks', fontsize = 18)
```
